```python
import yfinance
import pandas as pd

symbols = [
    'SPY',
    'AAPL',
    'MSFT',
    'AMZN',
    'GOOG',
    'NVDA',
]

dataframes = []

for symbol in symbols:
    ticker = yfinance.Ticker(symbol)
    df = ticker.history(period="max")
    df.reset_index(inplace=True)

    df['Ticker'] = symbol

    dataframes.append(df)

all_data = pd.concat(dataframes)
# all_data.to_csv('../data/stock_data.csv')
all_data.head()
```

```python
import duckdb

con = duckdb.connect(database='../data/data.duckdb', read_only=False)

con.sql('create or replace table history as select * from all_data')
```
